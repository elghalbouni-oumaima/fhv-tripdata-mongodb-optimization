### Type data

```python
dt_cols = ['request_datetime',	'pickup_datetime',	'dropoff_datetime','on_scene_datetime']
df[dt_cols] = df[dt_cols].apply(pd.to_datetime,errors='coerce')
```

```python
float_cols = ['trip_miles', 'base_passenger_fare', 'tolls', 
              'bcf', 'sales_tax', 'congestion_surcharge', 
              'airport_fee', 'tips', 'driver_pay']

df[float_cols]=df[float_cols].astype(float)
```

```python
int_cols = ['PULocationID', 'DOLocationID','trip_time']
df[int_cols] = df[int_cols].astype(int)
```

### Drop columns that we don't need

```python
df['access_a_ride_flag'].value_counts()
```

```python
df = df.drop(columns=['access_a_ride_flag'])
```

```python
df[['originating_base_num','on_scene_datetime']].isna().sum()
```

```python
df = df.drop(columns=['originating_base_num', 'on_scene_datetime'])
```

### Improving data format

```python
flag_cols = ['shared_request_flag','shared_match_flag', 'wav_request_flag',	'wav_match_flag']
df[flag_cols]= df[flag_cols].replace({'N':0,'Y':1})
```

```python
df['hvfhs_license_num'] = df['hvfhs_license_num'].str.upper().str.strip()
df['hvfhs_license_num'].head()
```

```python
df['dispatching_base_num'] = df['dispatching_base_num'].str.upper().str.strip()
df['dispatching_base_num'].head()
```
